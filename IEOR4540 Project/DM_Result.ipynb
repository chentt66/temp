```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot(df,title,y_label):
    plt.figure(figsize=(10, 5))  

    for column in df.columns[1:]:
        plt.plot(df['Epoch'], df[column], label=column)

    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel(y_label)

    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.show()
```

```python
ViL_train_accuracy = pd.read_excel('DM_Result.xlsx', sheet_name='Train_accuracy')[["Epoch","ViL_s","ViL_c"]]
plot(ViL_train_accuracy,"Training Accuracy of ViL from Scratch VS from ViT checkpoint", "Accuracy %")
```

```python
ViL_train_speed = pd.read_excel('DM_Result.xlsx', sheet_name='Train_speed')[["Epoch","ViL_s","ViL_c"]]
plot(ViL_train_speed,"Training Speed of ViL from Scratch VS from ViT checkpoint", "Speed seconds")
```

```python
def plot_hist(df, title, ylabel):
    plt.figure(figsize=(10, 5)) 
    bars = plt.bar(df.index, df[0], color='skyblue')

    plt.title(title)
    plt.xlabel('Model')
    plt.ylabel(ylabel)
    plt.xticks(rotation=45)

    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), 
                 ha='center', 
                 va='bottom',  
                 fontsize=8)   
        
    plt.show()
```

```python
test_accuracy = pd.read_excel('DM_Result.xlsx', sheet_name='Test_accuracy').T
test_accuracy
```

```python
plot_hist(test_accuracy,'Test Accuracy per Model', 'Accuracy %')
```
